```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

```python
plt.figure(figsize=(8,6))

plt.title('Gas Prices over Time (in USD)',fontdict={'fontweight':'bold',
                                                    'fontsize':'22'})

plt.plot(gas.Year, gas.USA,'b.-')
plt.plot(gas.Year, gas.Canada,'r.-')
plt.plot(gas.Year, gas['South Korea'],'g.-')
plt.plot(gas.Year,gas.Australia, 'y.-')

#Another way to plot many values

#countries_data = ['Australia', 'USA', 'Canada', 'South Korea']
#for country in gas:
    #if country in countries_data:
       # plt.plot(gas.Year, gas[country], marker='.')

plt.legend(['USA','Canada','South Korea','Austrialia'])

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

# Save graph as picture
#plt.savefig('Gas_price_figure.png', dpi=300)


plt.show()
```
